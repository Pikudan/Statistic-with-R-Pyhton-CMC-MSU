Сгенерировать данные из нормального распределения с различными параметрами и провести анализ с помощью графиков эмпирических функций распределений, квантилей, метода огибающих, а также стандартных процедур проверки гипотез о нормальности (критерии Колмогорова-Смирнова, Шапиро- Уилка, Андерсона-Дарлинга, Крамера фон Мизеса, Колмогорова-Смирнова в модификации Лиллиефорса и Шапиро-Франсия). Рассмотреть выборки малого (не более 50-100 элементов) и умеренного (1000-5000 наблюдений) объемов.

```r
ex_5 <- function(size){
    xs <- rnorm(size)
    #график эмпирической функции
    ep1<-mean(xs)
    ep2<-var(xs)
    pc = pnorm(xs,mean=ep1,sd=ep2)
    op <- par(mfrow=c(1,1), pty="s") 
    par(mfrow=c(1,3))
    mn <- paste(c("Эмпирическая КФР")) 

    plot(xs,pc, type="l",col="red", lwd=2, main=mn) 
    plot(ecdf(xs),add=TRUE)
    ggplot(mapping = aes(xs)) + stat_ecdf(geom = "step") 
    quantile(xs, c(0, 0.25, 0.5, 0.75, 1)) 
    library(car)
    #график квантилей
    qqnorm(xs)
    #график метода огибающих
    qqPlot(xs, dist= "norm", pch=1)     
    #стандартные процедуры проверки гипотез о нормальности     
    print(ks.test(xs, "pnorm"))
    print(shapiro.test(xs))
    install.packages('nortest') 
    library(nortest) 
    print(ad.test(xs))
    print(cvm.test(xs)) 
    print(lillie.test(xs))
    print(sf.test(xs))

}
```

```r
ex_5(100)
ex_5(1000)
ex_5(5000)
```

# 7. Продемонстрировать применение для проверки различных гипотез и различных доверительных уровней (0.9, 0.95, 0.99) следующих критериев:
a. Стьюдента, включая односторонние варианты, когда проверяемая нулевая гипотеза заключается в том, что одно из сравниваемых средних значений больше (или меньше) другого. Реализовать оценку мощности критериев при заданном объеме выборки или определения объема выборки для достижения заданной мощности;
b. Уилкоксона-Манна-Уитни (ранговые);
c. Фишера, Левене, Бартлетта, Флигнера-Килина (проверка
гипотез об однородности дисперсий).

```r
ex_7 <- function(df, sample1, sample2, size){
    #. Стьюдента, включая односторонние варианты, когда 
    #проверяемая нулевая гипотеза заключается в том, что 
    #одно из сравниваемых средних значений больше (или меньше) другого.

    #двусторонний и доверительный уровень - 0.9
    print(t.test(sample1, sample2, alternative="two.sided", paired = TRUE, conf.level = 0.9))
    #односторонний(больше) и доверительный уровень - 0.95
    print(t.test(sample1, sample2, alternative="greater", paired = TRUE, conf.level = 0.95))
    #односторонний(меньше)) и доверительный уровень - 0.99
    print(t.test(sample1, sample2, alternative="less" , paired = TRUE, conf.level = 0.99))

    #оценка мощности критериев 
    #n - заданный объем выборки 
    #delta – минимальная величина эффекта, которую мы хотим обнаружить в ходе эксперимента; 
    #sd – стандартное отклонение; 
    #sig.level – уровень значимости

    print(power.t.test(n = size, delta = 3.0, sd = 3, sig.level = 0.05))
    print(power.t.test(n = size, delta = 3.0, sd = 1.8, sig.level = 0.05,type  = "one.sample"))
    #Уилкоксона-Манна-Уитни (ранговые);
    print(wilcox.test(sample1, sample2, paired = T))
    #Фишера, Левене, Бартлетта, Флигнера-Килина (проверка гипотез об однородности дисперсий)
    library(car)
    print(var.test(Income ~ Gender, data=df))
    print(leveneTest(Income ~ Gender, data=df))
    print(bartlett.test(Income ~ Gender, data=df))
    print(fligner.test(Income ~ Gender, data=df))
}
```

toy - независимые выборки расходов женщин и мужчин
gamma1, gamma2 - зависимые углы в белке

```r
ex_7(toy, protein[0:500]$gamma1, protein[0:500]$gamma2, 20)
```

# 8. Исследование корреляционных взаимосвязей в данных с помощью коэффициентов корреляции Пирсона, Спирмена и Кендалла.


```r
ex_8 <- function(xs, ys){
    #"с помощью коэффициентов корреляции Пирсона
    print(cor.test(xs,ys))
    #с помощью коэффициентов корреляции Спирмена
    print(cor.test(xs, ys, method = "spearman"))
    #с помощью коэффициентов корреляции Кендалла
    print(cor.test(xs, ys, method = "kendall"))
}
```

# 11. Исследовать зависимости в данных с помощью дисперсионного анализа.

```r
toy_three_city = toy[toy$City %in%  c("Dallas" ,"Washington D.C." , "Austin")]
```

```r
ex_11 <- function(df, column1, column2){
#Однофакторный дисперсионный анализ
    anova <- aov(Income ~ City, data = df)
    plot(TukeyHSD(anova))
    print(aov(Income ~ City, data = df))
 
#Kruskal-Wallis test не предполагает ни нормальности популяции, 
#ни однородности дисперсии, как и параметрическая ANOVA, 
#и требует только упорядоченного масштабирования зависимой переменной. 
#Kruskal-Wallis используется, когда нарушения нормальности популяции 
#и однородности дисперсии являются экстремальными'''
    print(kruskal.test(Income ~ City, data = df))
}
```

# 12. Подогнать регрессионные модели (в том числе, нелинейные) к данным, а также оценить качество подобной аппроксимации.

```r
library(dplyr)
library(ggplot2)
```

Линейные модели

```r
ex_12_1 <- function(x, y){
    plot(x, y, pch=10, xlab = "psi", ylab = "tau")
    sapply(1:3,function(k) points(x,predict(lm(y~poly(x, k, raw = TRUE))), col=k, lwd=2 ))
    legend ("topleft", c("k = 1", "k = 2","k = 3"), col = 1:3, lwd = 2) 
    M_reg <- lm(y ~ x)
    print(summary(M_reg))
    
    M_reg2 <- lm(y ~ poly(x,2))
    print(summary(M_reg2))


    M_reg3 <- lm(y ~ poly(x,3))
    print(summary(M_reg3))
}
```

```r
ex_12_1(protein$psi, protein$tau)
```

```r
ex_12_2 <- function(X){
    set.seed(100)
    tr.index = sample(1: nrow(X), nrow(X) * 0.7)
    X_train <- X[tr.index]
    X_test <- X[tr.index]
    library(rpart)
    library(rpart.plot)
    fit <- rpart(City~., data = X_train, method = 'class')
    rpart.plot(fit)
    predict_unseen <-predict(fit, X_test, type = 'class')
    table_mat <- table(X_test$City, predict_unseen)
    accuracy_Test <- sum(diag(table_mat)) / sum(table_mat)
    print(paste('Accuracy for test', accuracy_Test))
}
```
